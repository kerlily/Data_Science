```python
import matplotlib.pyplot as plt

# Data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Membuat diagram batang
plt.bar(languages, popularity, color=['blue', 'orange', 'green', 'red', 'purple', 'brown'])
plt.title("Popularitas Bahasa Pemrograman")
plt.xlabel("Bahasa Pemrograman")
plt.ylabel("Popularitas (%)")
plt.show()
```

```python
# Membuat diagram batang horizontal
plt.barh(languages, popularity, color='skyblue')
plt.title("Popularitas Bahasa Pemrograman")
plt.xlabel("Popularitas (%)")
plt.ylabel("Bahasa Pemrograman")
plt.show()
```

```python
# Membuat diagram lingkaran
plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140)
plt.title("Popularitas Bahasa Pemrograman")
plt.show()
```

```python

import pandas as pd

# Membaca data dari file CSV
data = pd.read_csv("medal.csv")

# Membuat diagram lingkaran
plt.pie(data['gold_medal'], labels=data['country'], autopct='%1.1f%%', startangle=140)
plt.title("Pencapaian Medali Emas Olimpiade Musim Panas 2016")
plt.show()
```
