```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Data
X = np.array([1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]).reshape(-1, 1)
Y = np.array([2, 4, 5, 4, 6, 7, 8, 8, 10, 9])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Print model coefficients
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
print("Mean Squared Error (MSE):", mean_squared_error(Y, Y_pred))

# Plot
plt.scatter(X, Y, color='blue', label='Data')
plt.plot(X, Y_pred, color='red', label='Linear Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Model')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Membaca dataset
real_estate_data = pd.read_csv('real_estate.csv')

# Menampilkan kolom-kolom dataset
print("Kolom dataset:", real_estate_data.columns)

# Memilih variabel independen (X) dan dependen (Y)
X = real_estate_data['dist_to_mrt_m'].values.reshape(-1, 1)  # Jarak ke MRT
Y = real_estate_data['price_twd_msq'].values  # Harga per meter persegi

# Membuat model regresi linear
model = LinearRegression()
model.fit(X, Y)

# Melakukan prediksi
Y_pred = model.predict(X)

# Menghitung koefisien regresi, intercept, dan MSE
slope = model.coef_[0]
intercept = model.intercept_
mse = mean_squared_error(Y, Y_pred)

# Menampilkan hasil regresi
print(f"Koefisien (Slope): {slope}")
print(f"Intercept: {intercept}")
print(f"Mean Squared Error (MSE): {mse}")

# Membuat plot regresi
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('Distance to MRT (meters)')
plt.ylabel('Price per square meter (TWD)')
plt.title('Linear Regression: Distance to MRT vs Price')
plt.legend()
plt.show()
```
