```python
import matplotlib.pyplot as plt

# Membaca data dari file
x, y = [], []
with open('test1.txt', 'r') as file:
    for line in file:
        point = line.split()
        x.append(int(point[0]))
        y.append(int(point[1]))

# Plot garis
plt.plot(x, y, marker='o')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Garis dari Nilai Sumbu')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari CSV
data = pd.read_csv("fdata.csv")
data['Tanggal'] = pd.to_datetime(data['Tanggal'], format='%m-%d-%y')
data.set_index('Tanggal', inplace=True)

# Plot garis dengan grid
plt.plot(data.index, data['Tutup'], label='Penutupan')
plt.xlabel('Tanggal')
plt.ylabel('Harga Penutupan')
plt.title('Diagram Garis Data Keuangan Alphabet Inc.')

# Atur grid
plt.grid(visible=True, which='major', linestyle='-', linewidth=0.7, color='blue')
plt.grid(visible=True, which='minor', linestyle=':', linewidth=0.5, color='gray')
plt.minorticks_on()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data contoh
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 4, 6, 8, 10]

# Plot kedua garis
plt.plot(x, y1, label='Kuadrat', color='blue')
plt.plot(x, y2, label='Linear', color='green')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.title('Dua Garis pada Plot yang Sama')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data contoh
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 35]
y2 = [15, 25, 20, 30, 40]

# Plot dengan penanda
plt.plot(x, y1, label='Dataset 1', color='blue', marker='o', linestyle='-')
plt.plot(x, y2, label='Dataset 2', color='green', marker='s', linestyle='--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Dua Garis dengan Penanda pada Plot yang Sama')
plt.legend()
plt.show()
```
