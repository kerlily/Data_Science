```python
import matplotlib.pyplot as plt

# Membaca data dari file
x, y = [], []
with open('test1.txt', 'r') as file:
    for line in file:
        point = line.split()
        x.append(int(point[0]))
        y.append(int(point[1]))

# Plot garis
plt.plot(x, y, marker='o')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Garis dari Nilai Sumbu')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari CSV
data = pd.read_csv("fdata.csv")
data['Tanggal'] = pd.to_datetime(data['Tanggal'], format='%m-%d-%y')
data.set_index('Tanggal', inplace=True)

# Plot garis dengan grid
plt.plot(data.index, data['Tutup'], label='Penutupan')
plt.xlabel('Tanggal')
plt.ylabel('Harga Penutupan')
plt.title('Diagram Garis Data Keuangan Alphabet Inc.')

# Atur grid
plt.grid(visible=True, which='major', linestyle='-', linewidth=0.7, color='blue')
plt.grid(visible=True, which='minor', linestyle=':', linewidth=0.5, color='gray')
plt.minorticks_on()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

#data untuk garis pertama
x1 = [10, 20, 30]
y1 = [20, 40, 10]

#data untuk garis kedua
x2 = [10, 20, 30]
y2 = [40, 10, 30]

#membuat plot
plt.plot(x1, y1, label='Dataset 1', color='blue', alpha=0.5)
plt.plot(x2, y2, label='Dataset 2', color='green', alpha=0.5)

#menetapkan label dan judul
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Dua Garis dalam satu plot dengan legenda yang sesuai')

# mengatur ticks sumbu x dan y tidak deesimal
plt.xticks((10,20, 30))

#menampilkan legenda
plt.legend()

#menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data untuk garis pertama
x1 = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 13, 20]

# Data untuk garis kedua
x2 = [1, 2, 3, 4, 5]
y2 = [2, 4, 6, 8, 10]

# Data untuk garis ketiga
x3 = [1, 2, 3, 4, 5]
y3 = [5, 7, 9, 11, 13]

# Plot garis pertama
plt.plot(x1, y1, label='Garis 1', marker='o', linestyle='-', color='blue')

# Plot garis kedua
plt.plot(x2, y2, label='Garis 2', marker='s', linestyle='--', color='red')

# Plot garis ketiga
plt.plot(x3, y3, label='Garis 3', marker='^', linestyle='-.', color='green')

# Menambahkan judul dan label sumbu
plt.title('Plot Beberapa Garis dengan Penanda')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')

# Menampilkan legenda
plt.legend()

# Menampilkan plot
plt.show()

```
